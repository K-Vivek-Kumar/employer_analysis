```python
import pandas as pd
```

```python
path = "../Assignment_Timecard.xlsx - Sheet1.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head(4)
```

```python
from datetime import datetime, timedelta
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y %I:%M %p')
df['Time Out'] = pd.to_datetime(df['Time Out'], format='%m/%d/%Y %I:%M %p')
```

```python
grouped_data = df.groupby('Employee Name')
```

```python
def worked_for_7_days(group):
    sorted_dates = group['Time'].sort_values()
    return any((sorted_dates.iloc[i + 1] - sorted_dates.iloc[i]).days == 1 for i in range(len(sorted_dates) - 1))
```

```python
def less_than_10_hours_between_shifts(group):
    sorted_times = group['Time'].sort_values()
    time_diff = sorted_times.diff().fillna(pd.Timedelta(seconds=0))
    return any((1 < time_diff.iloc[i].seconds / 3600 < 10) for i in range(1, len(time_diff)))
```

```python
def more_than_14_hours_in_a_shift(group):
    shift_duration = (group['Time Out'] - group['Time']).dt.total_seconds() / 3600
    return any(shift_duration > 14)
```

```python
for name, group in grouped_data:
    if worked_for_7_days(group):
        print(f"{name} has worked for 7 consecutive days.")

    if less_than_10_hours_between_shifts(group):
        print(f"{name} has less than 10 hours but greater than 1 hour between shifts.")

    if more_than_14_hours_in_a_shift(group):
        print(f"{name} has worked for more than 14 hours in a single shift.")
```
